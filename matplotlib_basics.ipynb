########Generating Visualization with simple example

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([0,1,2,3,4])
plt.ylabel('some numbers')
plt.xlabel(' This is the x axis')
plt.show()
```

```python
plt.plot([1,2,3,4,0,9], [2,7,1,6,3,4])


plt.show()
```

```python
plt.plot([1,2,3,4,0,9], [2,7,1,6,3,4], 'ro')


plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['semester_1', 'semester_2', 'semester_3', 'semester_4', 'semester_5']
values = [3.24, 3.16, 3.00, 2.96, 3.5]

plt.figure(figsize=(16,4))

plt.subplot(151)
plt.bar(names, values)

plt.subplot(152)
plt.scatter(names, values)

plt.subplot(153)
plt.plot(names, values, linewidth = 2.0)

plt.subplot(154)
plt.stackplot(names, values)

plt.subplot(155)
plt.stem(names, values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
